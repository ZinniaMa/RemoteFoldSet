```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pickle
```

```python
from sklearn.manifold import TSNE
```

# ProtT5

```python
prot_df = pd.read_pickle("../embeddings/prott5_embeddings.pkl")
prot_df = prot_df.sort_values(by="name").reset_index(drop=True)
```

```python
X = np.vstack(prot_df['embedding'].values)
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, max_iter=1000, init='random', random_state=42)
X_2d = tsne.fit_transform(X)
```

```python
sns.set(style="darkgrid")

plt.figure(figsize=(4,4), dpi=300)
plt.scatter(X_2d[:,0], X_2d[:,1], s=10, alpha=0.7, color='#7C4DFF')
plt.title("t-SNE of ProtT5 Embeddings")
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()
```

# Sequences (one-hot encoded)

```python
path = "../saved/RemoteFoldSet_domain_to_seqs.pkl"
with open(path, "rb") as f:
    filename_to_seqs = pickle.load(f)
```

```python
all_seqs = []
for seqs in filename_to_seqs.values():
    all_seqs.extend(seqs)
all_seqs = [seq.replace('X', '') for seq in all_seqs]
len(all_seqs), all_seqs[:5]
```

```python
AAS = ""
for seq in all_seqs:
    for aa in seq:
        if aa not in AAS:
            AAS += aa
AAS = list(AAS)
print(AAS)
print(len(AAS))
```

```python
aa2idx = {aa:i for i, aa in enumerate(AAS)}
aa2idx
```

```python
def one_hot_encode(seq):
    mat = np.zeros((len(seq), len(AAS)), dtype=np.float32)
    for i, aa in enumerate(seq):
        if aa in aa2idx:
            mat[i, aa2idx[aa]] = 1
    return mat
```

```python
vecs = []
for seq in all_seqs:
    mat = one_hot_encode(seq)
    vecs.append(mat.mean(axis=0))
len(vecs), vecs[:3]
```

```python
X = np.vstack(vecs)
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, max_iter=1000, init='random', random_state=42)
X_2d = tsne.fit_transform(X)
```

```python
sns.set(style="darkgrid")

plt.figure(figsize=(4,4), dpi=300)
plt.scatter(X_2d[:,0], X_2d[:,1], s=10, alpha=0.7, color='black')
plt.title("t-SNE of sequences (one-hot encoded)")
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()
```
